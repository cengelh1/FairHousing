# Census Tract To Neighborhood Function
It takes census tract level geospatial data and dissolves it into neighborhood shapes. 
The arguments you need are your data frame (i.e. 'df= your data frame', your geometry column, usually named 'geometry' (i.e. "geometry='geometry'", and lastly how you want the variables to be summarized (i.e. 'sum' to add or 'mean' for an average)

```python
def nhdissolve(df,geometry,math,skip=False):
    
    #Import neighborhood shapes
    url = "https://usc.data.socrata.com/api/geospatial/r8qd-yxsr?method=export&format=GeoJSON"
    nhoods=gpd.read_file(url)
   
    #This is to make the geography column recognizable, toggling skip to True allows you to avoid an error message if your geometry column does not need this operation
    runwkts=skip
    if runwkts==False:
        df[geometry] = df[geometry].apply(wkt.loads)
    #add a CRS
    gdf = gpd.GeoDataFrame(df, crs="EPSG:4326",geometry='geometry')
     #join race with nh shapes, rename neighborhood column, and set it as the index
    nhjoin=gdf.overlay(nhoods, how='intersection')
    nhjoin=nhjoin.rename(columns={'name':'Neighborhood'})
    nhjoin.set_index('Neighborhood',inplace=True)
    
    #drop shapes not in city of LA
    nhjoin.drop(nhjoin[nhjoin['type'] != 'segment-of-a-city'].index, inplace = True)

    #approximate average for data in neighborhoods 
    nhdata= nhjoin.dissolve(by='Neighborhood', aggfunc= math)
    nhdata.head()
    return nhdata
```

```python
#Drop the columns that have no data
LACounty = LACounty.dropna(axis=1,how="all")
```

```python
LACounty.head()
```

```python
#Keep the columns that we want from the data set
columns_to_keep = ['Geo_FIPS',
                   'SE_B18002_001',
                   'SE_B18002_002',
                   'SE_B18002_003',
                   'Percent_Rent_Burden',
                   'Percent_Severe_Burden']
RentBurden = LACounty[columns_to_keep]
```

```python
#List the columns
columns = list(RentBurden)
columns
```

```python
#Change the column names
RentBurden.columns = ['FIPS',
'SE_B18002_001',
 'SE_B18002_002',
 'SE_B18002_003',
 'Percent_Rent_Burden',
 'Percent_Severe_Burden']
```

```python
#Import the geodata file for census tracts
LACountyTracts=gpd.read_file('data/Census_Tracts_2020.geojson')
```

```python
# Open the data and check to see what the columns are
LACountyTracts.head()
```

```python
LACountyTracts = LACountyTracts[['CT20','geometry']]
```

```python
# check to see if I have the desired subset 
LACountyTracts.head()
```

```python
#Add in the missing FIPS code information for the FIPS column
LACountyTracts['FIPS'] ='06' + '037' + LACountyTracts['CT20']
```

```python
#Check to see if the leading zero is back
LACountyTracts.head()
```

```python
#Merge together the data and geodata file on FIPS
RentBurdenTracts=LACountyTracts.merge(RentBurden,on="FIPS")
```

```python
#Check to see if the merge was successful
RentBurdenTracts.head()
```

```python
#Run the dissolve function to convert census tracts into their corresponding neighborhoods
RentBurdenNeighborhoods= nhdissolve(df=RentBurdenTracts,geometry='geometry',math='sum',skip=True)
```

```python
#Check to see the new data set with neighborhoods
RentBurdenNeighborhoods.head()
```

```python
#Add in column with percent rent burden
RentBurdenNeighborhoods['Percent_RentBurden']= 100 * (RentBurdenNeighborhoods['SE_B18002_002']/RentBurdenNeighborhoods['SE_B18002_001'])
```

```python
#Add in column with percent severe rent burden
RentBurdenNeighborhoods['Percent_SevereBurden']= 100 * (RentBurdenNeighborhoods['SE_B18002_003']/RentBurdenNeighborhoods['SE_B18002_001'])
```

```python
#check the data with the newly added percent columns
RentBurdenNeighborhoods.head()
```

```python
#Check to see the CRS 
RentBurdenNeighborhoods.crs
```

```python
#Convert the CRS to mercator projection
RentBurdenNeighborhoods_web_mercator = RentBurdenNeighborhoods.to_crs(epsg=3857)
```

```python
#Plot the data
fig, ax = plt.subplots(figsize=(20, 10))
RentBurdenNeighborhoods_web_mercator.plot(
                 column='Percent_RentBurden',
                 legend=True, 
                 scheme='quantiles',
                 ax=ax)

# turn the axes off
ax.axis('off')

# give it a title
ax.set_title('Percentage of Renter Households Paying 30% to 49% of Income on Rent')

# add a basemap
ctx.add_basemap(ax, crs=3857, source=ctx.providers.Stamen.TonerLite)
```

```python
#Plot the data
fig, ax = plt.subplots(figsize=(20, 10))
RentBurdenNeighborhoods_web_mercator.plot(
                 column='Percent_SevereBurden',
                 legend=True, 
                 scheme='quantiles',
                 ax=ax)

# turn the axes off
ax.axis('off')

# give it a title
ax.set_title('Percentage of Renter Households Paying 50% or more of Income on Rent')

# add a basemap
ctx.add_basemap(ax, crs=3857, source=ctx.providers.Stamen.TonerLite)
```
