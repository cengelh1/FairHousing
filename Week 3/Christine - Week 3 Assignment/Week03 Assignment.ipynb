```python
# Importing geopands to move on to mapping
import geopandas as gpd
```

```python
# reading geojson file with city of LA census tract boundaries 
tracts=gpd.read_file('data/LACity_Census_Tracts_2010_Population.geojson')
tracts.head()
```

```python
# plotting census tract data
tracts.plot(figsize=(12,10))
```

```python
# reviewing info about the dataset
tracts.info(verbose=True, show_counts=True)
```

```python
# creating subset of data with FIPS and geometry
tracts = tracts[['TRACTCE10','geometry']]
tracts.head()
```

```python
# making a new FIPS column to match social explorer data
tracts['FIPS'] ='06' + '037' + tracts['TRACTCE10']
tracts.head(10)
```

```python
# merging data in a new variable 
tracts_tenure=tracts.merge(df2,on="FIPS")
```

```python
tracts_tenure.head(10)
```

```python
# mapping number of renter-occupied households
tracts_tenure.plot(figsize=(12,10),
                  column='Occupied Housing Units: Renter Occupied',
                  legend=True,
                  scheme='NaturalBreaks')
```

```python
# This map tells us where high numbers of renter-occupied households are, but not how many there are relative to the total number of housing units.
```

```python
# mapping percent of renter-occupied households
tracts_tenure.plot(figsize=(12,10),
                  column='% Occupied Housing Units: Renter Occupied',
                  legend=True,
                  scheme='NaturalBreaks')
```

```python
# This map is more useful than the first in understanding the relative number and location of renter-occupied households.
```

```python
# I also want to look at the percent of owner-occupied households, which should be the inverse of the map above.
# mapping percent of owner-occupied households
tracts_tenure.plot(figsize=(12,10),
                  column='% Occupied Housing Units: Owner Occupied',
                  legend=True,
                  scheme='NaturalBreaks')
```
